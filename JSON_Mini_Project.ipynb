```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
json_df = pd.read_json('world_bank_projects.json')
json_df
```

### Find the 10 countries with most projects

```python
# create a Pandas Series that slices all rows from 'countrycode' and
# do a frequency count on the unique values in the column
countries = json_df.loc[:,'countryname'].value_counts()

# print top 10 (ince .value_counts() orders result in descending order)
countries.head(10)
```

### Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
with open('world_bank_projects.json') as json_file:
    json_data = json.load(json_file)

# create a dictionary to store code-name pairs of projects
code_name = {}

# populate dictionary with the project code-name pairs
for data in json_data:
    for d in data['mjtheme_namecode']:
        key = d['code']
        value = d['name']
        if (key not in code_name) & (value != ''):
            code_name[key] = value

# group the dataframe by 'code' and get frequency count by using count()
normalized_df = json_normalize(json_data, 'mjtheme_namecode')
all_themes = normalized_df.groupby(['code']).count().sort_values('name', ascending=False).reset_index('code')

# renaming labels
all_themes.columns = ['Code', 'Frequency']

# 'code' is not intuitive and doesn't let the end user know what the project is
# so map the grouped by dataframe with the correct project name (using 'code_name' dictionary)
all_themes['Theme Name'] = all_themes['Code'].map(code_name)
all_themes
```

### In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# sort the mjtheme_namecode dataframe by code
sorted_df = normalized_df.sort_values('code')

# replace all blank columns with NaN, then perform a forward fill
sorted_df.replace('', np.nan, regex=True).fillna(method='ffill')
```
