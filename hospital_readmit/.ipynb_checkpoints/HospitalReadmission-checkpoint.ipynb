# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

## B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

### 1. Setup an appropriate hypothesis test.

**Null hypothesis:** There is no correlation between the number of discharges and the readmission rates.

**Alternate hypothesis:** There is a strong correlation between the number of discharges and the readmission rates.

### 2. Compute and report the observed significance value (or p-value).

**Significance level:** 0.05

```python
# STORE NUMBER OF DISCHARGES AND READMISSION RATIO INTO ARRAYS
discharge = clean_hospital_read_df['Number of Discharges'][81:-3].values
readmission = clean_hospital_read_df['Excess Readmission Ratio'][81:-3].values
```

#### Using Scipy's pearsonr function

```python
scipy_r, scipy_p = stats.pearsonr(discharge, readmission)

print("Scipy's correlation coefficient:", scipy_r)
print("Scipy's p-value:", scipy_p)
```

The p-value that is obtained is extremely small, therefore the null hypothesis (no correlation between number of discharges and readmission rates) is rejected.

### *Other ways to perform hypothesis test*

#### Using critical value (hands on calculation)

```python
# GETTING SIZE OF SAMPLE
n = len(clean_hospital_read_df[81:-3])

# CALCULATE T-VALUE, P-VALUE, AND CRITICAL VALUE
t = scipy_r * math.sqrt((n - 2) / (1 - scipy_r**2))
critical_value = stats.t.ppf(0.95, df=n-2)
p = stats.t.sf(np.abs(t), n-2)*2

print('T-score:', t)
print('Critical value:', critical_value)
print('P-value:', p)

# QUESTION! I found online that there are two methods of determining significance via hypothesis test.
# 1.) Use the p-value
# 2.) Use the critical value. If r is NOT between the positive and negative critical value, then it is significant.
# Using the p-value, it is clearly smaller than the significance level, so we reject the null hypothesis 
# (and we can conclude r is significant), but r IS between the neg/pos critical value, so it isn't significant...
# I'm essentially getting two different answers. Am I doing something wrong?
```

#### Permute one set of data

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x, y)

    # Return entry [0,1]
    return corr_mat[0,1]
```

```python
# The one that was confusing me. I'm inclined to think I need to swap the inequality from '>=' to '<=' 
# because the r value is negative. Or maybe I can just use the absolute value of r. Is this approach 
# another alternative or is this wrong entirely?

perm_reps = np.empty(len(readmission))

for i in range(len(perm_reps)):
    discharge_permuted = np.random.permutation(discharge)
    perm_reps[i] = pearson_r(discharge_permuted, readmission)
    
pval = np.sum(perm_reps >= scipy_r) / len(perm_reps)

print('p-value:', pval)
```

### 3. Report statistical significance for  αα  = .01.

As the p-value is extremely small (1.5022756426464526 x 10<sup>-23</sup>), so despite changing the significance level from 0.05 to 0.01, the null hypothesis is still rejected.

### 4. Discuss statistical significance and practical significance. Do they differ here? How does this change your recommendation to the client?

Statistical significance refers to the likelihood that a certain statistic has occurred by chance. When performing a hypothesis test, we want to know that assuming the null hypothesis, how likely it would be to obtain a value (in the population) at least as extreme as what is observed in the sample. A small p-value usually leads us to reject the null hypothesis, which also tells us the value we performed a hypothesis testing on is statistically significant.

One thing to note about statistical significance is that as the sample size increases, so will the accuracy of our test. More specifically, our ability to detect minute differences becomes stronger. However, a statistically significant difference between what is observed and what is hypothesized does not necessarily mean the difference is practically significant.

Practical significance is how meaningful that a difference is in reality. Although by rejecting a null hypothesis we have found statistical significance, it may lack practical significance. Practicality is not calculable. It is something we determine based on our knowledge of what is being tested.

With that in mind, although there is a negative correlation between hospital capacity and readmission rates, the strength of the negative correlation seems rather weak. It might not be wise to solely focus on improving quality of care in smaller hospital facilities.

#### Joint plot

```python
ax = sns.jointplot(discharge, readmission, kind='hex')
ax.set_axis_labels(xlabel='Number of Discharges', ylabel='Excess Readmission Rate')
plt.show()
```

From our sample, we can see that hospitals that discharge less than 500 patients generally have readmission rates of 0.9-1.1%. 

```python
# SINCE 'clean_hospital_read_df['Number of Discharges']' IS SORTED, USE INDEX TO SLICE INTO 2 SERIES
# (capacity less than 500 = small)
small_hospital = clean_hospital_read_df['Number of Discharges'][81:8973]
big_hospital = clean_hospital_read_df['Number of Discharges'][8973:11578]

# CREATE NEW COLUMN 'category' AND ASSIGN APPROPRIATE HOSPITAL CAPACITY LEVEL TO DATA IN SAMPLE
small_hospital = small_hospital.to_frame().assign(category = 'small')
big_hospital = big_hospital.to_frame().assign(category = 'big')

# CONCATENATE FOR GRAPHING PURPOSES
hospital = small_hospital.append(big_hospital)
hospital = pd.concat([hospital, clean_hospital_read_df['Excess Readmission Ratio'][81:-3]], axis=1)

ax = sns.violinplot(x='category', y='Excess Readmission Ratio', data=hospital, hue='category', dodge=False, palette='RdBu')
ax.set_title('Violin plot of Number of Discharges vs Excess Readmission Rate')
plt.show()
```

Like the hex joint plot, we can see that a lot of facilities have an excess readmission ratio near 1%. The two groups of data have very similar shapes and visually, the means, interquartile ranges, and densities are very close. Admittedly, the tails of the violin plots clearly show that smaller hospital facilities have slightly more occurrences of higher readmission ratios. However, as mentioned above, the correlation between hospital capacity and readmission ratio is not practically significant.
